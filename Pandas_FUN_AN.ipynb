Задание 1
Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

```python
import pandas as pd
```

```python
ratings = pd.read_csv('ratings.csv')
ratings.head(5)
```

```python
def grade(i):
    if i <=2:
        return 'low'       
    if i <=4:
        return 'medium'
    if i >=4.5:
        return 'high'
    else:
        return 'false'
```

```python
ratings['class'] = ratings['rating'].apply(grade)
ratings.head()
```

Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’.

Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:

geo_data = {

'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}

Результат классификации запишите в отдельный столбец region.

```python
keywords = pd.read_csv('keywords.csv')
keywords.head(1)
```

```python
def geo_data(a):
    if a == 'москва' or a =='тула' or a =='ярославль':
        return 'Центр'
    if a == 'петербург' or a == 'псков'or a == 'мурманск':
        return 'Северо-Запад'
    if a == 'владивосток' or a == 'сахалин'or a == 'хабаровск':
        return 'Дальний Восток'
    else:
        return 'undefined'
```

```python
keywords['region'] = keywords['keyword'].apply(geo_data)
keywords.head()
#проверка
keywords_2 = keywords[(keywords['keyword'] == 'москва')|(keywords['keyword'] == 'петербург')|(keywords['keyword'] == 'владивосток')]
keywords_2
```

```python
def production_year(row):
    for i in years:
        if str(i) in row:
            return i
    else:
        return '1900'
```
