```python
import pandas as pd
```

```python
ratings = pd.read_csv('ratings.csv')
movies = pd.read_csv('movies.csv')
```

```python
ratings.head(2)
```

```python
movie_1 = ratings[ratings.rating == 5].groupby('movieId').count().sort_values('rating',ascending=False).head(1)
movie_2 = movie_1.merge(movies, on='movieId', how='left')
movie_3 = movie_2.filter(items = ['movieId', 'rating', 'title', 'genres'])
movie_3
```

```python
power = pd.read_csv('power.csv')
```

```python
power.describe()
```

```python
power_1 = power[ (power.quantity > 0 )]
power_2 = power_1[ (power_1.category == 4)| (power_1.category == 12) | (power_1.category == 21)]
power_2.describe()
```

```python
power_3 = power_2[ (power_2['country']== 'Latvia') | (power_2['country']=='Lithuania') | (power_2['country']=='Estonia')]
power_3[(power_3.year >= 2005) & (power_3.year <= 2010)].tail(2)
```

```python
power_3[(power_3.year >= 2005)&(power_3.year <= 2010)].groupby('category').sum().sort_values('category').head()
```
